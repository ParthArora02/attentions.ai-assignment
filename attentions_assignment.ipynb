load and combine json files

```python
import json
import os

# Paths to the directories
metrics_table_new_path = '/content/data_model/Attentions.ai Assignment/DataModel/metrics_table_new'
ecom_tables_and_columns_new_with_dimensions_path = '/content/data_model/Attentions.ai Assignment/DataModel/ecom_tables_and_columns_new_with_dimensions'

# Function to load all JSON files from a given directory
def load_json_files_from_directory(directory_path):
    json_data = []
    for filename in os.listdir(directory_path):
        if filename.endswith('.json'):
            with open(os.path.join(directory_path, filename), 'r') as f:
                data = json.load(f)
                json_data.append(data)
    return json_data

# Load JSON files from both directories
metrics_table_data = load_json_files_from_directory(metrics_table_new_path)
ecom_tables_data = load_json_files_from_directory(ecom_tables_and_columns_new_with_dimensions_path)

# Combine the data
combined_data_model = {
    "metrics_table": metrics_table_data,
    "ecom_tables_and_columns": ecom_tables_data,
    "relationships": []  # This will be filled with relationships from the minimized relationships file
}

# Check the combined data
print(json.dumps(combined_data_model, indent=4))
```

merge relationships


```python
# Path to the relationships file
relationships_file_path = '/content/data_model/Attentions.ai Assignment/relationships_minimized.json'

# Load relationships from the minimized relationships file
with open(relationships_file_path, 'r') as f:
    relationships_minimized = json.load(f)

# Add the relationships to the combined data model
for table, relations in relationships_minimized.items():
    for relation in relations:
        combined_data_model["relationships"].append(relation)

# Save the updated data model to a new JSON file
with open('/content/updated_data_model.json', 'w') as f:
    json.dump(combined_data_model, f, indent=4)
```

```python
from transformers import AutoTokenizer, AutoModel

# Load a model (e.g., distilbert-base-uncased)
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModel.from_pretrained("distilbert-base-uncased")

def get_embeddings(texts):
  encoded_inputs = tokenizer(texts, padding="max_length", truncation=True, return_tensors="pt")
  with torch.no_grad():
    outputs = model(**encoded_inputs)
  return outputs.last_hidden_state[:, 0, :].cpu().numpy()  # Extract last hidden state for first token

# Get embeddings for your texts
texts = [
    "This is the first text.",
    "This is another text for embedding.",
    "Yet another example text."
]
embeddings = get_embeddings(texts)

# Print the embeddings (assuming embeddings is a list of numpy arrays)
for text, embedding in zip(texts, embeddings):
  print(f"Text: {text}")
  print(f"Embedding: {embedding}\n")
```
